# 파일 불러오기

```python
import pandas as pd
```

```python
claim = pd.read_excel('1,2회 통합본2.xlsx')
```

```python
claim.head()
```

```python
claim2 = claim[['문장번호','단어']]
claim2.head()
```

```python
claim2.dtypes
```

```python
claim2['문장번호'] = claim2['문장번호'].astype('int')
```

```python
claim2['단어'] = claim2['단어'].astype('object')
```

```python
c_list = claim2.values.tolist()
c_list[:10]
```

```python
len(c_list)
```

```python
c_middle = []
c_middle.append(c_list[0][1])
c_middle
```

```python
c_list2 = []
for i in range(len(c_list)-1):
    if c_list[i][0] == c_list[i+1][0]:
        c_middle.append(c_list[i+1][1])
    else:
        c_list2.append(c_middle)
        c_middle=[]
        c_middle.append(c_list[i+1][1])
c_list2.append(c_middle)
```

```python
c_list2[:10]
```

```python
len(c_list2)
```

```python
for i in range(len(c_list2)):
    for j in range(len(c_list2[i])):
        c_list2[i][j] = str(c_list2[i][j])
```

```python
len(c_list2)
```

```python
from gensim.models import FastText

# 기존 실험(Word2vec을 사용한 패턴 뽑기)에서 사용한 파라미터
model2 = FastText(c_list2, size = 300, window = 15, min_count = 7, workers=4, iter=500, sg=1)
```

```python
model2.wv.vectors.shape
```

```python
model2.save('claim_all_fasttext2.model')
```
